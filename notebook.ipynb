# Count the number of short action movies from the 1990s

```python
netflix_df_short_action_movies_1990s = netflix_df_1990s[ (netflix_df_1990s['genre'] == 'Action') & (netflix_df_1990s['duration']<90)]
```

```python
# Count the number of short action movies released in the 1990s and save this integer as 'short_movie_count'.
short_movie_count = 0
for index, row in netflix_df_short_action_movies_1990s.iterrows():
    short_movie_count += 1

print(short_movie_count)    
```
